```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
```

```python
data = pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,accuracy_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(data.iloc[:,0:-1],data.iloc[:,-1])
```

```python
def preprocess(x,y):
    enc = OneHotEncoder()
    y_out = enc.fit_transform(y.values.reshape(-1,1))
    scaler = StandardScaler()
    x_out = scaler.fit_transform(x)
    return x_out,y_out.toarray()
```

```python
xin,yin = preprocess(x_train,y_train)
```

```python
xt,yt = preprocess(x_test,y_test)
```

```python
def print_report(y_true,y_pred):
    print(classification_report(y_true,y_pred))
    print(f"Accuracy:- {accuracy_score(y_true,y_pred)}")
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(4,)))
model.add(tf.keras.layers.Dense(3,activation='softmax'))
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),loss=tf.keras.losses.categorical_crossentropy)
```

```python
model.fit(xin,yin,epochs=100,validation_split=0.1,verbose=1)
```

```python
pred = model.predict(xt)
```

```python
print_report(yt,(pred > 0.5))
```
